### **Developers Survey Response Time.**

```python
time_start=pd.to_datetime(df['StartDate'])
time_end=pd.to_datetime(df['EndDate'])
t_time = {'Begin': time_start, 'End': time_end}
df_time=pd.DataFrame(t_time)
#let's find the time difference using timedelta
l=list()
for i,j in zip(df_time['End'],df_time['Begin']):
    l.append(pd.Timedelta(i-j).seconds / 60.0)
print('The Time Taken by Respondents are as follows: \n',l)
```

```python
print('The Minimum Time Taken by any single Respondents is: ',min(l),' minutes')
print('The Maximum Time Taken by any single Respondents is: ',max(l),' minutes')
print('The Average Time Taken by any single Respondents is: ',sum(l)/len(l),' minutes')
```

```python
#let's create a series of the list.
s=pd.Series(l)
s
```

```python
(s>100).value_counts()
#this means 8081 people answered survery in 15 minutes.
```

```python
plt.figure(figsize=(15,6))
s.hist(bins=500,edgecolor='red', linewidth=1.2)
plt.xlim([0,100])
plt.xticks(range(0,100,5))
plt.title('Developers Response Time Distribution')
plt.show()
```
